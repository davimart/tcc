```python
import pandas as pd
import json
from pathlib import Path
from sklearn.model_selection import train_test_split

# ============================================================
# Caminhos
# ============================================================

pasta_tratado = Path("datasets/tratado")
raw = Path("datasets/raw")
caminho_hmda_original = raw / "hmda_100.csv"
caminho_hmda_5pct = pasta_tratado / "hmda_15pct.csv"

caminho_configs = Path("configs_datasets.json")

# ============================================================
# Carrega config para pegar o target correto
# ============================================================

with open(caminho_configs, "r", encoding="utf-8") as f:
    configs = json.load(f)

target_col = configs["hmda"]["target"]

# ============================================================
# Carrega HMDA original
# ============================================================

df_hmda = pd.read_csv(caminho_hmda_original)

n_original = len(df_hmda)

print(f"Linhas no HMDA original: {n_original:,}".replace(",", "."))

# ============================================================
# Amostra 5% estratificada pelo target
# ============================================================

df_hmda_5pct, _ = train_test_split(
    df_hmda,
    train_size=0.15,
    random_state=42,
    stratify=df_hmda[target_col]
)

df_hmda_5pct = df_hmda_5pct.reset_index(drop=True)

n_5pct = len(df_hmda_5pct)

print(f"Linhas no HMDA 5%: {n_5pct:,}".replace(",", "."))
print(f"Percentual usado: {100 * n_5pct / n_original:.2f}%")

# ============================================================
# Conferência da distribuição do target
# ============================================================

print("\nDistribuição do target no dataset original:")
print(df_hmda[target_col].value_counts(normalize=True))

print("\nDistribuição do target no dataset 5%:")
print(df_hmda_5pct[target_col].value_counts(normalize=True))

# ============================================================
# Salva novo arquivo
# ============================================================

df_hmda_5pct.to_csv(caminho_hmda_5pct, index=False)

print(f"\nArquivo salvo em: {caminho_hmda_5pct}")
```
